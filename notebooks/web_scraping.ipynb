# 1. Extracting URL of Cities

### Reading Dineout URL 

```python
r = requests.get('https://www.dineout.co.in/')
soup = bs(r.content)
# print(soup.prettify())
```

### Extracting city-wise URLs 

```python
from urllib.parse import urljoin
home_url = 'https://www.dineout.co.in'
page_url = '-restaurants/welcome-back?p=' 
# cities_list = pd.DataFrame()
cities_url = pd.DataFrame()

for city in soup.find(class_='_3LAV0'):
#     print(city)
#     print(city.text)
    cities = city.find('a',href=True)
#     cities_list = cities_list.append({'Cities Links':cities['href']}, ignore_index =  True)
#     print(cities['href'])
    cities_url = cities_url.append({'URL': urljoin(home_url, (cities['href'] + page_url))}, ignore_index = True)
```

```python
# cities_url.head(20)
```

### Extracting page-wise URLs for every city

```python
pagewise_urls = pd.DataFrame()

for index, link in enumerate(cities_url['URL']):
    i = 1 
    a=1

    while a is not None:
    #             print(link)
            pagewise_url = link+str(i)
        #         url = 'https://www.dineout.co.in/goa-restaurants/welcome-back?p={}'.format(i)
            r = requests.get(pagewise_url)
            soup = bs(r.content, 'html.parser')
            print(pagewise_url)
            i += 1
            a = soup.find(class_='restnt-detail-wrap')

            if a is not None:
                pagewise_urls = pagewise_urls.append({'URL': pagewise_url}, ignore_index = True)
```

```python
# from urllib.parse import urljoin
restnt_url = pd.DataFrame()

for index, link in enumerate(pagewise_urls['URL']):
    r = requests.get(pagewise_urls['URL'][index])
    
    print(pagewise_urls['URL'][index])

    # Convert to a beautifulsoup object
    soup = bs(r.content)
    rest_list = soup.find_all(class_='restnt-name ellipsis')
    
    for restnt in rest_list:
        restnt_link = restnt.get('href')
        review_link = '/review'
        restnt_url = restnt_url.append({'Link' : urljoin(home_url, (restnt_link + review_link))}, ignore_index=True)
```

# 3. Saving Restaurant Details

### Declaring Function to Obtain Restaurant Details

```python
# rest_details = pd.DataFrame()
def get_restnt_info(url):

    r = requests.get(url)
    soup = bs(r.content)
    
    rest_details = {}
    rest_info = soup.find(class_='restnt-infoBox')
    rest_details['Name'] = rest_info.find(class_='restnt-name').get_text()
    rest_details['Rating'] = rest_info.find(class_='restnt-rating-widget').get_text()    
    rest_details['Location'] = rest_info.find(class_='restnt-loc').get_text()
    rest_details['Cuisine'] = rest_info.find(class_='cuisine-type').get_text()
#     rest_details['Offers'] = rest_info.find(class_='links-tab')('li')[2].get_text()
    rest_details['Votes'] = rest_info.find(class_='rating-txt')
    if not rest_details['Votes']:
        rest_details['Votes'] = 0
    else:
        rest_details['Votes'] = rest_info.find(class_='rating-txt').get_text()
    rest_details['Cost'] = rest_price = rest_info.find(class_='fs16 marginB5').get_text()
    return rest_details
```
